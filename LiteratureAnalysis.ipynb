```python
a = open("D:\\Python\\rich.txt","r")
b = open("D:\\Python\\gen.txt","r")
business = a.read()
novel= b.read()
```

Here's a function to calculate the average length of sentences in text. We need to tokenize sentences for further analysis. 

```python
def Sentences(text):
    tokenizer = nltk.sent_tokenize(text)
    average_s = sum(len(i) for i in tokenizer) / len(tokenizer)
    print("The average length of sentences in this book is ", average_s)

print("First book")
Sentences(business)
print("Second book")
Sentences(novel)
```

```python
def Words(text):
    tokenizer = nltk.RegexpTokenizer(r"\w+")
    new_words = tokenizer.tokenize(text)
    print("There are ", len(new_words), " words in the text.")
    average_w = sum(len(i) for i in new_words) / len(new_words)
    print("The average length of words is", average_w, ".")

print("First book")
Words(business)
print("Second book")
Words(novel)
```

Now we have to work with words, so here it's important to unify the text, for that we will remove symbols and we will write everything in lower case letters.

```python
def FilterSymbols(text):
    a = text
    symb = [",",".","(",")","[","]","\"","?",";",":",".","!"]
    for i in symb:
        a=a.replace(i,"")
        return a

business_filtered = FilterSymbols(business).lower()
novel_filtered = FilterSymbols(novel).lower()
```

Tokenizing words, deleting "stopwords".

```python
def cleaning(txt):
    text_tokens = word_tokenize(txt)
    tokens_without_sw = [word for word in text_tokens if not word in stopwords.words()]
    return tokens_without_sw

a = cleaning(business_filtered)
b = cleaning(novel_filtered)
```

Now we have a version of a simple text, let's see the frequency of words. That can be best visualized with word cloud. We will run this on a different version of our texts as it takes forever to generate a wordcloud in a different way.

```python
def plot_cloud(wordcloud):
    plt.figure(figsize=(40, 30))
    plt.imshow(wordcloud) 
    plt.axis("off");
```

```python
wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, background_color='black', 
                      colormap='Set2', collocations=False, stopwords = STOPWORDS).generate(business_filtered)

plot_cloud(wordcloud)

wordcloud.to_file("wordcloud.png")
```

```python
wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, background_color='salmon', 
                      colormap='Pastel2', collocations=False, stopwords = STOPWORDS).generate(novel_filtered)

plot_cloud(wordcloud)
```

```python
from nltk.book import *
from nltk.text import Text 

bus = Text(nltk.corpus.gutenberg.words("D:\\Python\\rich.txt"))
bus.concordance("money",100)
```

```python
genius = Text(nltk.corpus.gutenberg.words("D:\\Python\\gen.txt"))
genius.concordance("money",100)
```

```python
bible = Text(nltk.corpus.gutenberg.words("bible-kjv.txt"))
bible.concordance("money",100)
len(bible)
```

Now we have 3 books, let's create a list to save time. <br> Checking if the word "money" is used in the same context

```python
books=[bus, genius, bible]

for i in books:
    i.similar("money")
    print("\n")
```

```python
for i in books:
    i.similar("rich")
    print("\n")
```

```python
for i in books:
    i.common_contexts(["money"])
    print("\n")
```

```python
for i in books:
    i.dispersion_plot(["money", "rich", "work", "poor", "education", "rest", "family", "balance"])
```

```python
for i in books:
    i.dispersion_plot(["good","happy","amazing","beautiful","pleasant","wonderful","up","bad", "sad","annoying","cheat","anxiety","fear","terrible","upset","down"])
```

Well, as we paid attention at the word "money", let's see the proportion of this word in the text (we'll use the formula for all of the books).

```python
for i in books:
    print("Frequency of the word: ",i.count("money"))
    a = 100 * i.count("money") / len(i)
    print("Proportion: ",a, "%")
```

```python
def Frequency(text):
    fdist = FreqDist(text)
    print(fdist)
    b = fdist.tabulate(20)
    fdist.plot(50, cumulative=True)

for i in books:
    Frequency(i)
```

```python
fdist1 = FreqDist(bus)
fdist2 = FreqDist(genius)
fdist3 = FreqDist(bible)
a = sorted(w for w in set(fdist1) if len(w) > 6 and fdist1[w] > 10)
b = sorted(w for w in set(fdist2) if len(w) > 6 and fdist2[w] > 10)
c = sorted(w for w in set(fdist3) if len(w) > 6 and fdist3[w] > 10)

d={"Business Book":[a],"Genius": [b],"Bible": [c]}
df=pd.DataFrame(d)
df
```
